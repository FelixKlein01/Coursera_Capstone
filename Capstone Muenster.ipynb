## Get geolocation and cafe Data

```python
# Get Muenster geolocation to be used for centering the map
address = 'Muenster'
geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude

print(latitude, longitude)
```

```python
area = 'Muenster'
section = 'food'
category = '4bf58dd8d48988d16d941735'    # cafe category id
limit = 100

url = 'https://api.foursquare.com/v2/venues/explore?\
client_id={}&\
client_secret={}&\
near={}&v={}&\
section={}&\
categoryId={}&\
limit={}'\
.format(CLIENT_ID, CLIENT_SECRET, area, VERSION, section, category, limit)

url
```

```python


results = requests.get(url).json()
venues = results['response']['groups'][0]['items']
df = pd.json_normalize(venues)

df

```

## Cleaning Data

```python

        
# Reindex rows after clean up
df.reset_index(drop=True, inplace=True)

# Rename columns
df.rename(columns = {'venue.id':'id', 'venue.name':'name', 'venue.location.formattedAddress':'address', 'venue.location.lat':'lat', 'venue.location.lng':'lng'}, inplace = True) 

# Select only columns of interest
dff = df[['id', 'name', 'address', 'lat', 'lng']]

dff
```

Unfortunally the Data about these cafes is not complete. There are no Ratings aviable. To solve this problem the google maps data could be useful. To continue with this project we add radom ratings.

```python
dff['score'] = np.random.randint(0, 500, dff.shape[0]) / 100

dff
```

## Results of Cafe search

```python
# generate map centered to our location
venues_map = folium.Map(location=[latitude, longitude], zoom_start=14)

# add venues
for i, row in dff.iterrows():

    strength = round(row['score'] / 5, 2)
    if strength > 0.8:
        mcolor = "Crimson"
    else:
        mcolor = "DarkSlateBlue"

    folium.CircleMarker(
        [row['lat'], row['lng']],
        radius = 15,
        color = "#CCC",
        popup = str(i) + ': ' + str(row['score']),
        fill = True,
        fill_color = mcolor,
        fill_opacity = strength,
        opacity = 1
    ).add_to(venues_map)

# add a red marker for Muenster
folium.CircleMarker(
    [latitude, longitude],
    radius = 1,
    color = 'white',
    fill = True,
    fill_color = 'red',
    fill_opacity = 1
).add_to(venues_map)
        
venues_map
```
